### Challenge 1

```python
def bins_function(x):
    if 0<x<1000:
        i = 1
    if 1000<=x<9999:
        i = 2
    if 10000<=x<99999:
        i = 3
    if 100000<=x<999999:
        i = 4
    if x > 1000000:
        i = 5
    return i

bins_function(999)
```

```python
df["bin"] = df["Votes"].apply(bins_function)
```

```python
df
```

```python
def  rev_function(x, y):
    z = x*y
    return z

df["rev_min"] = df.apply(lambda x: rev_function( x['Runtime (Minutes)'], x['Revenue (Millions)']), axis=1 )
```

### Challenge 3

```python
def gen_function(rank, gen):
    if 'Comedy' in gen:
        return rank-1
    if 'Thriller' in gen:
            return rank+1
    else:
        return rank
```

```python
gen_function(2, "Thriller")
```

```python
df["Rank_2"] = df.apply(lambda x: gen_function(x["Rank"], x["Genre"]), axis=1 )
```

```python
df
```

### Challenge 4

```python
def prime_num(t, v):
    d = len(t)    
    r = int(d/v)
    if r > 1:
        for i in range(2, int(v)):
            if int((r % i)) == 0:
                return False
        return True
    else:
        return False
```

```python
df["prime"] = df.apply(lambda x: prime_num(x["Title"], x["Votes"]), axis=1 )
df
```
